#### 检查数据中是否有空值

```python
# 统计空值的数量，根据数量大小排序
train.isnull().sum().sort_values(ascending=True)
```

```python
# 统计空值的数量，根据数量大小排序
test.isnull().sum().sort_values(ascending=True)
```

```python
# 删除掉train中为空的数据
train.drop(train[train.isnull().any(1)].index, axis=0, inplace=True)
```

#### 检查车费fare_amount这一列

```python
train['fare_amount'].describe()
```

```python
# 统计train中车费小于0的数据有多少
from collections import Counter

Counter(train['fare_amount'] < 0)
```

```python
# 删除掉车费小于0的数据
train.drop(train[train['fare_amount'] < 0].index, axis=0, inplace=True)
```

```python
# 查看车费数据
train['fare_amount'].describe()
```

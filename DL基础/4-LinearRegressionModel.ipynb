## Terminology

Training set: Data used to train the model

Notation:
- $x$ = "input"(feature) variable
- $y$ = "output"(target) variable
- $m$ = number of training examples
- $(x, y)$ = single training example
- $(x^{(i)}, y^{(i)})$ = $i^{th}$ training example, where $i \in \{1, 2, ..., m\}$, and $i$ refers to a specific row in the table
  - e.g. $x^{(1)} = 2104$

| size in feet^2(x)     | price in $1000's(y)          |
| ----------------- | --------------------- |
| 2104   |  400  |
| 1416   |  232  |
| 1534   |  315  |
| 852    |  178  |
| ...    |  ...  |
| 3210   |  870  |


## How to represent $f$

Let's insist that $f$ is a straight line, so the function can be written as:
$$
f_{w, b}(x)=wx+b
$$

Linear regression with one variable = Univariate linear regression

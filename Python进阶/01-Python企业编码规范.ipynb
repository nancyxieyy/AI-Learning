技巧 6： EAFP 优于 LBYL(请求原谅比获得许可容易)

```python
d = {'x': 5}
try:
    value = int(d['x'])
    print(value)
except (KeyError, TypeError, ValueError):
    value = None
```

技巧 7：使用 enumeration 进行跌倒

```python
fruits = ['apple', 'banana', 'orange', 'grape', 'pear']
for index, fruit in enumerate(fruits):
    print(index, ':', fruit)
```

技巧 8：用生成式生成列表、集合和字典

```python
data = [1, 2, 3, 4, 5]  
result = [num * 3 for num in data if num > 10]
print(result)
```

技巧 9：用 zip 组合键值来创建字典

```python
keys = ['1001', '1002', '1003']
values = ['Alice', 'Bob', 'Charlie']
d = dict(zip(keys, values))
print(d)
```

## 案例三：文本生成(LSTM)

### 1. 下载文本、解析

```python
import tensorflow.keras
import numpy as np

path = tensorflow.keras.utils.get_file(
    'nietzsche.txt',
    origin = 'https://s3.amazonaws.com/text-datasets/nietzsche.txt'
)
# 读取文件，并转为小写
text = open(path).read().lower()
# 输出：文本长度
print("文本长度：%s"%len(text))
```

### 2. 将字符序列向量化

```python
maxlen = 60 # 提取60个字符组成的序列
step = 3 # 每隔3个字符提取一次新序列
sequences = [] # 保存所有序列
labels = [] # 保存所有标签
# 构建数据集
for i in range(0, len(text)-maxlen, step):
    sequences.append(text[i : i+maxlen]) # 数据
    labels.append(text[i + maxlen]) # 标签

print('sequences中的元素个数：%d'%len(sequences))
print('labels中的元素个数：%d'%len(labels))
```

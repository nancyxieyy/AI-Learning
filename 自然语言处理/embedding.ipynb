word embedding(词嵌入)
- 广义的认为word embedding包括所有密集词汇向量的表示方法，比如word2vec是word embedding的一种
- 狭义上是指在神经网络中加入的embedding层，对整个网络进行训练的同时产生的embedding矩阵(embedding层的参数)，这个embedding矩阵就是训练过程中所有输入词汇的向量表示组成的矩阵

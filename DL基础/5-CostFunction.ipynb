# Cost function: Squared error cost function

$J(w, b) = \frac{1}{2m} \sum_{i=1}^{m} (\hat{y}^{(i)} - y^{(i)})^2$

error: $\hat{y}^{(i)} - y^{(i)}$

m = number of training examples

$\hat{y}^{(i)}$ = $f_{w, b}(x^{(i)})$
